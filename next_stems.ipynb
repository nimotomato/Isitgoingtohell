```python
# 4. Compute mean label score for each date and region (Calculate_ratio_dated) with pandas. Upload to Table 3.
from isitgoingtohell.data_management.data_analysis import Dated_methods as CRD
from isitgoingtohell.data_management.db_management import DB
from isitgoingtohell.utils import load_csv
import pandas as pd
import plotly.express as px


db = DB()
crd = CRD()
country_codes = load_csv("only_codes.csv")
data = db.get_all_data(tablename="data")
date_score_region = crd.to_dict(crd.calculate_ratio_dated(), not_null=True)
regions = [
    'africa',
    'asia',
    'europe',
    'oceania',
    'north america',
    'south america'
]
date_score_region

def populate_countries(country_codes, date_score_region, region):
    min_nya_lista = []

    for country in country_codes:
        if country['region'].lower() == region:
            for score in date_score_region:
                if score['region'] == region:
                    nytt_dict_item = {}
                    nytt_dict_item['country_code'] = country['code']
                    nytt_dict_item['dated_region_score'] = score['score']
                    nytt_dict_item['date'] = score['date']
                    nytt_dict_item['region'] = score['region']
                    min_nya_lista.append(nytt_dict_item)
    return min_nya_lista

new_list = []
for region in regions:
    new_list.extend(populate_countries(country_codes, date_score_region, region))

fig = px.choropleth(new_list, locationmode="ISO-3", 
    locations="country_code",
    color="dated_region_score",
    hover_name="region",
    title = "News sentiment plotted using Plotly", 
    animation_frame= "date",
    color_continuous_scale=px.colors.sequential.PuRd)
 
 
fig["layout"].pop("updatemenus")
fig.show()
        
```
